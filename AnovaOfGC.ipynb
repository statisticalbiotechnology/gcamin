```python
import pandas as pd

df = pd.read_csv("df_uniprot_reference_proteomes_per_species_dec2017.csv")
```

```python
df
```

```python
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm

aa_dict = {}
for aa_col in [col for col in df.columns if 'freq_' in col]:
    lm = smf.ols( aa_col + ' ~ C(kingdom)+ GC', data=df).fit()
    table = sm.stats.anova_lm(lm, typ=2)
    aa_dict[aa_col] = table["F"]["C(kingdom)"]
    print(aa_col)
    print(table)
```

```python
fv = pd.DataFrame()
fv = fv.append(aa_dict,ignore_index=True).T.rename(columns={0: "F"})
fv.sort_values(by="F",ascending=False,inplace=True)
```
